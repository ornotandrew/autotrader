```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('btcusdt.csv')
time_columns = ['Open time', 'Close time']
df[time_columns] = df[time_columns].apply(pd.to_datetime)
```

```python
from autotrader.backtest import backtest
from autotrader.strategy.random import handle_tick as random
from autotrader.strategy.constants import Signal

history = backtest(df, random, 1000)
```

```python
fig, ax1 = plt.subplots()

history.plot.line(y="Base balance", ax=ax1)
ax2 = ax1.twinx()
history.plot.line(y="Counter balance", ax=ax2, color='red')

plt.show()
```
